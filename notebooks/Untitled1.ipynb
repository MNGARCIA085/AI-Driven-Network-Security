```python
df = pd.read_csv("../data/output2.csv")
```

```python
# Remove leading/trailing spaces from all column names
df.columns = df.columns.str.strip()
```

```python
# remove extra rare cases
# Keep only classes with >= 2 samples


df_sample = df

counts = df_sample['Label'].value_counts()
valid_classes = counts[counts >= 2].index
df_sample = df_sample[df_sample['Label'].isin(valid_classes)]


"""
or combine rare attacks into rare attacks
rare_threshold = 50  # or choose small number
counts = df_sample['Label'].value_counts()
rare_classes = counts[counts < rare_threshold].index

df_sample['Label'] = df_sample['Label'].replace(rare_classes, 'Other_Attack')
"""
```

```python
# separate in train/val df and test df
from sklearn.model_selection import train_test_split


data = df_sample


# Suppose df_sample has 100k rows
train_val_df, test_df = train_test_split(
    data,
    test_size=0.2,  # 20% for test
    stratify=data['Label'],
    random_state=42
)
```
